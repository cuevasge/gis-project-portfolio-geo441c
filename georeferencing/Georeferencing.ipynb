### **2. Learn about Steps in Georeferencing**

- Watch [this video](https://youtu.be/J-to3kHo1Ug?feature=shared) to learn how to georeference a map in ArcGIS Pro
- Esri official help document for [Georeferencing tool](https://pro.arcgis.com/en/pro-app/latest/help/data/imagery/georeferencing-tools.htm)
- Start selecting ground control points (GCPs) and try to minimize the Root Mean Squared Error (RMSE) of the georeferenced map
  - **Select 6-10 GCPs** to apply georeferencing and control the overall errors
- Rules of selecting GCPs
  - **Select a fixed object visible in both the image to be georeferenced and the basemap, for example**
    - Street/block corners
    - Building corners (be aware of building shape change across years)
    - Street intersections
  - Try to spread the GCPs as evenly distributed across the entire map as possible
    - **DO NOT select your first three GCPs along the same linear structure**
- Click Save button to apply georeferencing to the historical map
- You can compare how much difference there is between then and now on our campus and nearby areas in Oxford
- Once you are done with georeferencing, click **Generate Report** button to calculate the Total RMSE
  - **Copy and paste your forward and inverse RMSE to [this Google Sheet](https://docs.google.com/spreadsheets/d/1WonAP_gQL_Ra_4R1gZiIZ-vqZ8EkM7Eyyf2160KwCHw/edit?usp=sharing)**
