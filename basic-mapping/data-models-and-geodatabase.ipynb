- **Write a 100-word paragraph to describe the relationship between**
  - **`Geodatabase`**
  - **`Feature Dataset`**
  - **`Feature Classes`**
  - I can refer to **Page 9** of [assignment guide](https://miamioh.instructure.com/courses/246872/files/folder/week-02-resources?preview=38195670) or other resources online, such as
    - [Fundamentals of geodatabase](https://pro.arcgis.com/en/pro-app/latest/help/data/geodatabases/overview/fundamentals-of-the-geodatabase.htm)
    - [Feature datasets in ArcGIS Pro](https://pro.arcgis.com/en/pro-app/latest/help/data/feature-datasets/feature-datasets-in-arcgis-pro.htm)
- The geodatabase is the mechanism that holds our feature dataset and our feature classes in an organized manner. A feature class is a table that stores geometric properties of geographic features. Each row corresponds to one feature and contains a reference to its point, line, and polygon attributes. The geodatabase stores a collection of feature datasets, which each contain a set of feature classes that share the same coordinate system. Feature datasets allow us to organize our data and use more advanced functions, such as Topology and Terrain data. Our geodatabase is how we begin to organize our GIS designs.
