### **Reflection on my first experience with ArcGIS Pro**

- Overall, I spent 2 hours to finish creating my first map
- Here is a table showing things that I feel excited and challenging in my first experience working in ArcGIS Pro

| Things I'm excited about      | Things challenging for me  |
| ----------------------------- | -------------------------- |
| Getting familiar with geospatial data!    | Learning how to navigate the files  |
| Focusing on really specific aspects of a certain location    | Making things the right size    |
| Learning how to make things visually appealing      | Connecting my folder to the catalog   |

- Here is how I rate my first experience working in ArcGIS Pro
  - Data importing and management
    - At first I had a really hard time putting my assignement data in the right place to access from my project. I kept moving it across different folders and was not sure where it was supposed to stay so that I could access it from multiple projects. Then, after some practice on my first two maps, I was able to successfully add a folder connection between my Cloquet map and the Assignment 1 data.
  - Map creation and annotation
    - When creating my map, I ran into some problems with sizing. On the Hugo map, I was trying to resize my map to make it fit better on the page, but kept cutting off pieces of my data, which was then showing an incomplete picture. Later on, I found the "Activate" function, which allowed me to choose a size to make my map on the layout, and then adjust the zoom and positioning without moving the borders.
  - Navigate the software using its GUI
    - For my first map, I had trouble navigating the GUI to find the symbology tab, and was looking for it in the wrong places. Once I got the hang of it, I was able to tell more intuitively where I would have to find a command on the top ribbon, or right click on an element to find its properties or individual settings.
